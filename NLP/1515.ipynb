```python
"""A module with dictionaries.

Lab-08 : Dictionaries and Sorting with keys.
2 key functions and 1 main function.

key function - get_fre(pair)
               get the frequncy from the 2-tuple; (character, frequency)
               for the key to sort.

             - get_char(pair)
               get the character from the 2-tuple; (character, frequency)
               for the key to sort.

main function - sort_freq(filename)
                get the information of the given file.
                This function returns the 3 informations in form of
                3-tuple; (filesize, num_chars, dlist).
                    - filesize; size of the file.
                    - num_chars; the number of unique characters in file.
                    - dlist; the frequencies of the each characters
                      in the form of tuple in sorted list.
"""


def get_fre(pair):
    """
    The function for the key of sort the list
    of the 2-tuple; (character, frequency), by the frequencies.

    : param pair : 2-tuple; (character, frequencies)
    : param type : tuple

    : returns : frequencies
    : return type : int

    : examples : get_fre(('a', 2)) is 2
                 get_fre(('b', 8)) is 8

    : preconditions :   the type of pair is tuple.
    """

    # assert the precondition
    assert type(pair) == tuple

    # return the frequencies
    return pair[1]


def get_char(pair):
    """
    The function for the key of sort the list
    of the 2-tuple; (character, frequency), by the frequencies.

    : param pair : 2-tuple; (character, frequencies)
    : param type : tuple

    : returns : character
    : return type : str

    : examples : get_char(('a', 2)) is 'a'
                 get_char(('b', 8)) is 'b'

    : preconditions :   type of pair is tuple
    """

    # assert the precondition
    assert type(pair) == tuple

    # return the character
    return pair[0]


def sort_freq(filename):
    """
    This function is programmed to get the information of the given file.
    This function returns the 3 informations
    - file size, the number of unique characters in file and, the frequencies
    of the each characters in the form of tuple in sorted list.
    The list of the tuple; (character, frequencies) should be sorted
    in the increasing frequencies, if the frequencies is same, then, should be
    sorted in the order of increasing ord(character).

    : param filename : name of file from which the information for the returns
                       will be extracted.
    : type filename : str

    :returns: 3-tuple; (filesize, num_chars, dlist),
              where
              filesize: size of the ``filename`` as total number of characters
                        read from file.
              num_chars: number of unique characters in the file.
              dlist: list of 2-tuples, [(c1, n1), (c2, n2), (c3, n3), ...],
                     where
                     c1, c2, ... = each character in the given file,
                     n1, n2, ... = frequencies of occurrences of the
                                   characters.
                     in increasing frequencies, if same in order of
                     increasing ord(character).
    :return type: 3-tuple of (int, int, list)
    """

    # assert the precondition
    assert type(filename) == str

    # open the file
    file = open(filename, 'r', encoding='utf-8', newline='')

    # declare the variables
    filesize = 0
    dic = {}

    # read lines in file
    for line in file:
        # count the file size
        filesize += len(line)
        # update the dictionary of character:frequency
        for cha in line:
            if cha not in dic:
                dic[cha] = 1
            else:
                dic[cha] += 1

    # count the number of unique characters in file
    num_chars = len(dic)

    # make dictionary into the list
    d_list = []
    for char in dic:
        d_list += [(char, dic[char])]

    # sort the list in order
    d_list.sort(key=get_char)
    d_list.sort(key=get_fre)

    # close the file
    file.close()

    return filesize, num_chars, d_list


if __name__ == '__main__':
    # If you need codes to test your functions above, you shall write
    # your test code here inside this if-statement. Or, you may add
    # a test function and call the function here inside this if-statement.
    # Codes written below do "not" affect execution for grading since
    # professor and TA import this module and call only the above functions
    # and classes for grading.
    # But, the codes written below "do" affect grading of code styles since
    # all the codes in this file are used for grading purposes.
    # You're welcome to delete this if-statement block if this is not used,
    # and there should be no printings to console when this module is imported.
    pass

# -- end of 2020310105.윤정빈.HW8.py
```
